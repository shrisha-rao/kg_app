```python
# Reference to a collection named "users" (collection is created when you add a doc)
users_col = db.collection("test_users2")
```

```python
# Add a document with auto-generated ID
doc_ref = users_col.add({
    "name": "Alice",
    "email": "alice@example.com",
    "created_at": firestore.SERVER_TIMESTAMP
})
print("Added document:", doc_ref)
```

```python
# Or create a document with a specific ID
user_id = "alice-123"
users_col.document(user_id).set({
    "name": "Alice",
    "email": "alice@example.com",
    "created_at": firestore.SERVER_TIMESTAMP
})
print("Set document with ID:", user_id)
```

```python
# Read documents from the collection
for doc in users_col.stream():
    print(doc.id, doc.to_dict())
```

```python
from google.cloud import firestore
import os

# --- Configuration Check ---
# This line ensures the client connects to your emulator
# If FIRESTORE_EMULATOR_HOST is set, it will connect to the emulator.
# If not, it will connect to a real database (if credentials are available).
print(f"FIRESTORE_EMULATOR_HOST: {os.environ.get('FIRESTORE_EMULATOR_HOST')}")

# Initialize the Firestore client.
# Based on your UserService, this should auto-configure to the emulator.
try:
    db = firestore.Client(project="dev-project") # Use the project ID from your docker-compose
    print("Successfully initialized Firestore client.")
except Exception as e:
    print(f"Error initializing Firestore client: {e}")
    # Handle connection errors if necessary
    
# --- Inspection ---

# A. List all collections (Optional, for general inspection)
print("\n--- Collections ---")
collections = list(db.collections())
collection_names = [c.id for c in collections]
print(f"Found Collections: {collection_names}")


# B. Inspect the 'users' collection and find the test user (Crucial for debugging)
collection_name = "test_users3"
if collection_name in collection_names:
    print(f"\n--- Contents of '{collection_name}' Collection ---")
    
    # Get all documents in the 'users' collection
    users_ref = db.collection(collection_name)
    docs = users_ref.stream()

    found_test_user = False
    
    for doc in docs:
        user_id = doc.id
        user_data = doc.to_dict()
        print(f"  User ID: {user_id}")
        
        # Check if the specific test user exists
        if user_id == "test_user_123":
            found_test_user = True
            print("  >>> FOUND TARGET USER: 'test_user_123' <<<")
            import json
            print(f"  Data: {json.dumps(user_data, indent=2)}")

    if not found_test_user:
        print("\n*** ERROR DEBUG: The test user 'test_user_123' does NOT exist. ***")
        print("Please create this document in the 'users' collection.")

else:
    print(f"Error: Collection '{collection_name}' not found.")
```
